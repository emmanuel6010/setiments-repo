# Text transformation

```python
# copy data into df
df = data.copy()
df.sample(1)
```

```python
# inspect the customer review contents
df.review_content.head(2)
```

```python
# function to change a text column to lower case
def change_to_lower(dataframe, column):
    dataframe[column] = dataframe[column].astype(str).str.lower()
    return dataframe
```

```python
# change review content to lower case
df = change_to_lower(df, 'review_content')

# inspect dataframe
df.review_content.sample(2)
```

# Tokenization
- Breaking text into individual words

```python
# import mudule
from nltk.tokenize import RegexpTokenizer
```

```python
# function to tokenize a column
def tokenization(dataframe, column):
    regexp = RegexpTokenizer('\w+')
    dataframe['tokenized_review'] = dataframe[column].apply(regexp.tokenize)
    return dataframe
```

```python
df = tokenization(df, 'review_content')
```

```python
df.head()
```

## Stopwords
- Words which are generally filtered out before processing

```python
import nltk
nltk.download('stopwords')
```

```python
# get english stop words
stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(['https', 'jpeg', 'jpg', 'mp3', 'mp4'])
stopwords[:10]
```

```python
# remove stopwords from tokenized_review
df['tokenized_review'] = df['tokenized_review'].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
df.sample(1)
```

## Remove infrequent words
- Only words with more than two letters should be kept.

```python
# change the data type of tokenized_review to strings
# text_string to contain strings with words greater than 1
df['review_string'] = df['tokenized_review'].apply(lambda x: ' '.join([item for item in x if len(item)>1]))
```

```python
df.sample(2)
```

```python
# combine individual text into one list
overall_words = ' '.join([word for word in df['review_string']])

# tokenize overall_words
tokenize_overall_words = nltk.tokenize.word_tokenize(overall_words)
```

```python
tokenize_overall_words[:7]
```

# Word frequency count

```python
from nltk.probability import FreqDist
```

```python
word_frequency = FreqDist(tokenize_overall_words)
word_frequency
```

```python
# drop words which occur less than 1 time
df['review_string_frequency'] = df['tokenized_review'].apply(lambda x: ' '.join([item for item in x if word_frequency[item] >= 1 ]))
```

```python
df.sample(2)
```

## Lemmatization
- Break words down to their root meaning to identify similarities

```python
# check module update
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
# create lemmatizer object
lemmatizer = WordNetLemmatizer()

# lemmatize review_string_frequency
df['review_string_lemmatized'] = df['review_string_frequency'].apply(lemmatizer.lemmatize)
```

```python
df.sample(2)
```

# Sentiment analysis
### Using VADER (Valence Aware Dictionary and Sentiment Reasoner)

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# create a SentimentIntensityAnalyzer object
analyzer = SentimentIntensityAnalyzer()
```

```python
# polarity scores
df['polarity'] = df['review_string_lemmatized'].apply(lambda x: analyzer.polarity_scores(x))
```

```python
df.sample(2)
```

```python
# transform dataframe
df = pd.concat(
    [df.drop(['polarity'], axis=1),
    df['polarity'].apply(pd.Series)
    ], axis=1
)
```

```python
df.sample(20)
```

```python
# Create new variable with sentiment "neutral," "positive" and "negative"
df['sentiment_review_content'] = df['compound'].apply(lambda x: 'positive' if x >0 else 'neutral' if x==0 else 'negative')
```

```python
df.head()
```

### Visualize sentiment counts

```python
# count number of individual sentiment reviews
sns.countplot(y='sentiment_review_content', 
             data=df, 
             palette=['#b2d8d8',"#008080", '#db3d13']
             );
```
